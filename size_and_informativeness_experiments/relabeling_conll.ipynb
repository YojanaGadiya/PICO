```python
import nltk
import re
from nltk.tokenize import word_tokenize, sent_tokenize
from tqdm import tqdm
import pickle

#with open('kMeans_data/train_k=10_percentage=10.txt') as file:
with open('kMeans_data/train_200_uncurated.txt') as file:
    data = file.readlines()
```

```python
dir_name = 'kMeans_data'
fout = open(dir_name + '/relabeled_train_200_uncurated.txt', 'w')
#fout = open(dir_name + '/relabeled_train.txt', 'w')
list_of_lines = []
for line in data:
    list_of_lines.append(line)
    #print(len(line))


for i in range(len(list_of_lines)):
    #print(list_of_lines[i+1])
    line = list_of_lines[i]
    if line.startswith('-DOC'):
        continue
        fout.write(line)
        fout.write('\n')
        #print(doc_id)
    elif len(line)==1:
        #print(line)
        fout.write(line)
        #fout.write('\n')
    else:
        print(line, len(line))
        tok_line = line.split(" ")
        try: 
            if bool(re.search('mg', list_of_lines[i+1])):

                #print(line, list_of_lines[i+1])
                line = ' '.join([tok_line[0], tok_line[1], 'N'])
                fout.write(line)
                fout.write('\n')
            elif bool(re.search('mg', line)):
                line = ' '.join([tok_line[0], tok_line[1], 'N'])
                fout.write(line)
                fout.write('\n')
                #print(line)
            elif bool(re.search('patients', line)):
                line = ' '.join([tok_line[0], tok_line[1], 'p'])
                fout.write(line)
                fout.write('\n')
            elif bool(re.search('[de|in]crease', list_of_lines[i-1])) and list_of_lines[i-1].split(" ")[1]=='VBD':
                #print(line, list_of_lines[i+1])
                line = ' '.join([tok_line[0], tok_line[1], 'o'])
                fout.write(line)
                fout.write('\n')
            elif bool(re.search('[de|in]crease', line)):
                #print(line)
                line = ' '.join([tok_line[0], tok_line[1], 'o'])
                fout.write(line)
                fout.write('\n')
            elif bool(re.search('therapy', line)):
                line = ' '.join([tok_line[0], tok_line[1], 'i'])
                fout.write(line)
                fout.write('\n')
            elif bool(re.search('(otic|xel|xine|mab|phen|rall|lol|cillin|olam|acin|pam|eine|tin|zide|ax|oft|cin|done|lyn) ', line)):
                line = ' '.join([tok_line[0], tok_line[1], 'i'])
                fout.write(line)
                fout.write('\n')
            elif bool(re.search('(day|daily|weekly|) ', line)):
                line = ' '.join([tok_line[0], tok_line[1], 'N'])
                fout.write(line)
                fout.write('\n')
            elif bool(re.search('placebo', line)):
                line = ' '.join([tok_line[0], tok_line[1], 'N'])
                fout.write(line)
                fout.write('\n')

            else:
                line = line
                #print(line)
                fout.write(line)
                fout.write('\n')
        except IndexError:
            fout.write(line)
            #print("index error")
            continue
            
            
            
            
            
            
#(xel|xine|mab|phen|rall|lol|cillin|olam|acin|ram|pam|eine|tin|zide|ax|oft|cin|ex|done|lyn)$
```
